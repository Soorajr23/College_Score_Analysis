```python
# Getting Dept. Wise Total Sum
dept_sum = r1.pivot_table(index = 'Department', values = 'CodeKata Score', aggfunc = 'sum')
```

```python
dept_sum
```

```python
data = dept_sum['CodeKata Score']
labels = ['CSE', 'ECE', 'EEE']
explode = (0.2,0,0)
color = sns.color_palette('bright')
```

```python
# Plotting a PIE chart to display the sum of CodeKata Dept. Wise
fig = plt.figure(figsize = (10,5))
plt.pie(data, labels = labels, explode = explode, colors = color, autopct = '%1.1f%%')
plt.title("Total CodeKata Score Dept. Wise")
plt.show()
```

```python
# Getting Max marks in department
dept_max = r1.pivot_table(index = 'Department', values = ['Name','CodeKata Score'], aggfunc = 'max')
```

```python
dept_max
```

```python
# Plotting a Bar Plot of the dept. wise max results
fig = plt.figure(figsize = (10,6))
x = ['CSE', 'ECE', 'EEE']
y = dept_max['CodeKata Score']
plt.bar(x,y, width = 0.5)
plt.title("Highest Score per Dept")
plt.xlabel("Departments")
plt.ylabel("CodeKata Score")
plt.show()
```
